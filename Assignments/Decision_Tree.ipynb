## **Step 1: Load libraries.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import numpy as np
import pandas as pd
```

## **Step 2: Load data**

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

### **Visualización de los datos**

```python
# Convertir los datos a un DataFrame para mejor visualización
df = pd.DataFrame(X, columns=data.feature_names)
# Agregar la columna objetivo 'target'
df['target'] = y
```

```python
# Mostrar las primeras filas del DataFrame
df.head()
```

```python
df.describe()
```

## **Step 3: Data Preprocessing**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## ***Step* 4: Model Selection and training**

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
# Exactitud del modelo en el entrenamiento.
train_accuracy = clf.score(X_train, y_train)
print(train_accuracy)
```

##**Step 5: Model testing**

```python
y_pred = clf.predict(X_test)
```

```python
#Prediction 1
print(y_pred[1])
print(y_test[1])
```

```python
#Prediction 2
print(y_pred[20])
print(y_test[20])
```

```python
#Prediction 3
print(y_pred[-1])
print(y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names = data.target_names))
```

**Decision tree visualization.**

```python
plt.figure(figsize=(20, 15))  # Increase figure size for better resolution
tree.plot_tree(clf, filled=True, feature_names=data.feature_names, class_names=data.target_names)
plt.title("Decision Tree, Breast Cancer Wisconsin (Diagnostic)")
plt.savefig("decision_tree.png", dpi=300)  # Increase DPI for better resolution
plt.show()
```
